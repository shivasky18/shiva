```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,precision_score,recall_score,f1_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/credit_data.zip')
print(data.head(5))
```

```python
print(data.tail())
```

```python
data.shape
data.info()
```

```python
data.isnull().sum()
```

```python
print(data['Class'].value_counts())
print(f"{0} is non fraud trancation and {1} is fraud tranctions")
```

```python
legit=data[data.Class==0]
fraud=data[data.Class==1]
print(legit.head())
print(fraud.head())
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
data.groupby('Class').mean()
```

```python
legit_data=legit.sample(n=492)
# print(legit_data.head())
new_data=pd.concat([legit_data,fraud],axis=0)
print(new_data.head())
print(new_data.shape)
print(new_data.tail())
# print(new_data['Class'].value_counts())
```

```python
new_data.groupby('Class').mean()
```

```python
x=new_data.drop('Class',axis=1)
y=new_data['Class']
print(x.head())
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,stratify=y,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(y_pred)
```

```python
acc1=accuracy_score(y_test,y_pred)
print("yoo our model testing accuracy is : ",acc1*100)
```

```python
pred_train=model.predict(x_train)
# print(pred_train)
acc2=accuracy_score(y_train,pred_train)
print("yoo our model traing accuracy is : ",acc2*100)
```

```python
cls_report=classification_report(y_test,y_pred)
print(cls_report)
```

```python
con_matrix=confusion_matrix(y_test,y_pred)
print(con_matrix)
```

```python
pre_score=precision_score(y_test,y_pred)
print(pre_score)
```

```python
recall_score=recall_score(y_test,y_pred)
print(recall_score)
```

```python
f1_score=f1_score(y_test,y_pred)
print(f1_score)
```
